```python
cleaned_price = autos["price"].str.replace("$","").str.replace(",","").astype(int)
cleaned_odometer = autos["odometer"].str.replace("km","").str.replace(",","").astype(int)
autos["price"] = cleaned_price
autos["odometer"] = cleaned_odometer
print(autos["price"].dtype)

autos.rename({"odometer":"odometer_km"}, axis=1, inplace=True)
print(autos["odometer_km"].dtype)
```

```python
five_percent_and_more = autos["brand"].value_counts(normalize=True,ascending=False) > 0.005


counter = 0
top_brands = []
for s in five_percent_and_more:
    if s:
        top_brands.append(five_percent_and_more.index[counter])
    counter += 1

print(len(top_brands))
```

```python
average_price_of_top_brands = {}

for brand in top_brands:
    avg_price = autos.loc[autos["brand"]==brand,"price"].mean()
    
    if brand not in average_price_of_top_brands:
        average_price_of_top_brands[brand] = round(avg_price,2)
        
for k,v in average_price_of_top_brands.items():
    print(k + ": " + str(v))
```

```python
average_mileage_of_top_cars = {}

for brand in top_brands:
    avg_mileage = autos.loc[autos["brand"]==brand,"odometer_km"].mean()
    
    if brand not in average_mileage_of_top_cars:
        average_mileage_of_top_cars[brand] = round(avg_mileage,2)
```

```python
bma_series = pd.Series(average_mileage_of_top_cars)
bmp_series = pd.Series(average_price_of_top_brands)
```

```python
df = pd.DataFrame(bma_series,columns=["mean_mileage"])
df["mean_price"] = bmp_series
```
